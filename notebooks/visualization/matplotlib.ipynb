# MatPlotLib

The foundational plotting library for Python and Pandas



```python
import matplotlib.pyplot as plt  
import numpy as np 
```

```python
t = np.arange(0.0, 2.0, 0.01)  
s = np.sin(2*np.pi*t)  
plt.plot(t, s)  

plt.xlabel('time (s)')  
plt.ylabel('voltage (mV)')  
plt.title('voltage (mV) vs. time (sec)')   
plt.grid(True)  
#plt.savefig("plot-voltage-vs.-time.png")  
plt.show()  
```
