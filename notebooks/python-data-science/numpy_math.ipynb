### Classical Trigonometric Functions
> (x, /, out=None, *, where=True, casting='same_kind', order='K', dtype=None, subok=True[, signature, extobj])
>
> - np.**sin** [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.sin.html)

> - np.**tan** [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.tan.html)

```python
p.figure(figsize=(16, 8))

subhead("sin, cos")
x = np.linspace(0,10*np.pi,200)
ysin = np.sin(x)
ycos = np.cos(x)
ycosec = np.reciprocal(ysin, where=ysin!=0)

pycode("x = np.linspace(0,20*np.pi,100)")
pycode("ysin = np.sin(x)")
pycode("ycos = np.cos(x)")

p.plot(x, ysin, color="blueviolet", label='sin')
p.plot(x, ycos, color="dodgerblue", label='cos')
p.xlim(0, 10*np.pi)


p.grid(True, which='both', color='lavender')
p.axhline(y=0, color='lightsteelblue')
p.axvline(x=0, color='lightsteelblue')

p.legend(loc='lower right')
display(p)
```

```python
p.figure(figsize=(16, 8))

subhead("cosecant, secant")
x = np.linspace(0,10*np.pi,1000)
ysin = np.sin(x)
ycos = np.cos(x)
ycosec = np.reciprocal(ysin, where=ysin!=0)
ysecant = np.reciprocal(ycos, where=ycos!=0)

# remove points outside the view range
ycosec[np.abs(ycosec)>20] = np.nan
ysecant[np.abs(ysecant)>20] = np.nan

pycode("x = np.linspace(0,20*np.pi,100)")
pycode("ysin = np.sin(x)")
pycode("ycos = np.cos(x)")
pycode("ycosec = np.reciprocal(ysin, where=ysin!=0)")
pycode("ysecant = np.reciprocal(ycos, where=ycos!=0)")

p.plot(x, ysin, color="darkgoldenrod", label='sin')
p.plot(x, ycos, color="firebrick", label='cos')
p.plot(x, ycosec, color="khaki", label='cosecant')
p.plot(x, ysecant, color="lightcoral", label='secant')
p.ylim(-10, 10)
p.xlim(0, 10*np.pi)

p.grid(True, which='both', color='linen')
p.axhline(y=0, color='peachpuff')
p.axvline(x=0, color='peachpuff')

p.legend(loc='lower right')
display(p)
```

```python
p.figure(figsize=(16, 8))

subhead("tan")
x = np.linspace(0,10*np.pi,1000)
ytan = np.tan(x) # equivalent to np.sin(x) / np.cos(x)
ysin = np.sin(x)
ycot = np.divide(np.cos(x), ysin, where=ysin!=0)

# remove the -ve infinty - +ve infinity lines
ytan[:-1][np.diff(ytan) < 0] = np.nan
ycot[:-1][np.diff(ycot) >0] = np.nan

# write code
pycode("x = np.linspace(0,10*np.pi,1000)")
pycode("ytan = np.tan(x)")
pycode("ycot = np.divide(np.cos(x), ysin, where=ysin!=0)")

p.plot(x, ytan, color="darkorange", label='tan')
p.plot(x, ycot, color="gold", label='cot')
p.ylim(-10, 10)
p.xlim(0, 10*np.pi)

p.legend(loc='lower right')
display(p)
```

> **Inverse Trigonometry**

```python
p.figure(figsize=(10, 10))

subhead("arcsin, arccos")
x = np.linspace(-1*np.pi,np.pi,1000)

# create an x equivalent for arc functions, their domain is [-1,1]
arcx = x.copy()
arcx[(arcx<-1.0) | (arcx>1.0)] = np.nan

ysin = np.sin(x)
ycos = np.cos(x)
yarcsin = np.arcsin(arcx)
yarccos = np.arccos(arcx)

pycode("x = np.linspace(0,20*np.pi,100)")
pycode("ysin = np.sin(x)")
pycode("ycos = np.cos(x)")
pycode("yarcsin = np.arcsin(arcx)")
pycode("yarccos = np.arccos(arcx)")

p.plot(x, ysin, color = 'tomato', label="sin")
p.plot(x, yarcsin, color = 'lightsalmon', label="arc sin")
p.plot(x, ycos, color = 'royalblue', label="cos")
p.plot(x, yarccos, color = 'cornflowerblue', label="arc cos")
p.plot(x,x, dashes= [4,4], alpha=0.5, color='grey', label="x=y divider")

p.ylim(-1*np.pi, np.pi)
p.xlim(-1*np.pi, np.pi)

p.grid(True, which='both')
p.axhline(y=0, color='k')
p.axvline(x=0, color='k')

p.legend(loc='lower right')

display(p)
```

```python
p.figure(figsize=(12, 8))

subhead("arctan, arctan2")
x = np.linspace(-1*np.pi,np.pi,1000)
yarctan = np.arctan(x)
yarctan2 = np.arctan2(x, 0.3)

pycode("x = np.linspace(0,20*np.pi,1000)")
pycode("yarctan = np.arctan(x)")
pycode("yarctan2 = np.arctan2(x, 2)")
p.plot(x, yarctan, color = 'cornflowerblue', label='arctan')
p.plot(x, yarctan2, color = 'royalblue', label='arctan2, by 0.3')

p.ylim(-2, 2)
p.xlim(-5, 5)

p.grid(True, which='both')
p.axhline(y=0, color='k')
p.axvline(x=0, color='k')

p.legend(loc='lower right')

display(p)
```

## Hyperbolic functions

### Hyperbolic

> **Hyperbolic**
> (x, /, out=None, *, where=True, casting='same_kind', order='K', dtype=None, subok=True[, signature, extobj])
>
> - np.**sinh** [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.sinh.html)
> - np.**cosh** [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.cosh.html)
> - np.**tanh** [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.tanh.html)

### Inverse Hyperbolic
> (x, /, out=None, *, where=True, casting='same_kind', order='K', dtype=None, subok=True[, signature, extobj])
>
> - np.**arcsinh** [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.arcsinh.html)
> - np.**arccosh** [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.arccosh.html)
> - np.**arctanh** [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.arctanh.html)

```python
p.figure(figsize=(12, 8))

subhead("hyperbolic & inverse hyperbolic")
x = np.linspace(-10*np.pi,10*np.pi,1000)
sinh = np.sinh(x)
cosh = np.cosh(x)
tanh = np.tanh(x)

arcsinh = np.arcsinh(x)

xarccosh = x.copy()
xarccosh[xarccosh<1.0] = np.nan
arccosh = np.arccosh(xarccosh)

xarctanh = x.copy()
xarctanh[(x>-1) | (x<1)] = np.nan
arctanh = np.arctanh(xarctanh)

pycode("x = np.linspace(-10*np.pi,10*np.pi,1000)")
pycode("sinh = np.sinh(x)")
pycode("cosh = np.cosh(x)")
pycode("tanh = np.tanh(x)")
pycode("arcsinh = np.arcsinh(x)")
pycode("arccosh = np.arccosh(x)") 
pycode("arctanh = np.arctanh(x)")

p.plot(x, sinh, color = 'cornflowerblue', label="sinh")
p.plot(x, cosh, color = 'royalblue', label="cosh")
p.plot(x, tanh, color = 'skyblue', label="tanh")
p.plot(x, arcsinh, color = 'lightseagreen', label="arcsinh")
p.plot(xarccosh, arccosh, color = 'limegreen', label="arccosh")
p.plot(xarctanh, arctanh, color = 'teal', label="arctanh")
p.plot(x,x, dashes= [4,4], alpha=0.5, color='grey', label="x=y divider")

p.xlim(-2*np.pi, 2*np.pi)
p.ylim(-2*np.pi, 2*np.pi)

p.grid(True, which='both')
p.axhline(y=0, color='k')
p.axvline(x=0, color='k')

p.legend(loc='lower right')

display(p)
```

## Rounding

### Rounding
> (x, /, out=None, *, where=True, casting='same_kind', order='K', dtype=None, subok=True[, signature, extobj]) 
>
> - np.**round_** / np.**around**(a, decimals=0, out=None) [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.around.html)
> - np.**ceil** [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ceil.html) ceiling value
> - np.**floor** [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.floor.html) floor value
> - np.**trunc** [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.trunc.html) truncate float decimals
> - np.**fix** [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.fix.html) like truncate, round towards 0, float output
> - np.**rint**(a, decimals=0, out=None) [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.rint.html), round to nearest int

```python
p.figure(figsize=(12, 8))

subhead('Rounding')

x = np.linspace(-1*np.pi,np.pi,1000)
ysin = np.sin(x)*6
ysin_around = np.around(ysin,0)
ysin_rint = np.rint(ysin)
ysin_ceil = np.ceil(ysin)
ysin_floor = np.floor(ysin)
ysin_trunc = np.trunc(ysin)


pycode("x = np.linspace(-1*np.pi,np.pi,1000)")
pycode("ysin = np.sin(x)*6")
pycode("ysin_around = np.around(ysin,0)")
pycode("ysin_rint = np.rint(ysin)")
pycode("ysin_ceil = np.ceil(ysin)")
pycode("ysin_floor = np.floor(ysin)") 
pycode("ysin_trunc = np.trunc(ysin)")

p.rc('axes', prop_cycle=(cycler('color', ['lightslategrey', 'teal', 'limegreen', 'royalblue', 'darkgoldenrod', 'red'])))
p.plot(x, ysin, linewidth=1, label="ysin")
p.plot(x, ysin_around, dashes=[8, 8], label="around/round_")
p.plot(x, ysin_rint, dashes=[4, 5], label="rint")
p.plot(x, ysin_ceil, dashes=[4, 8],label="ceil")
p.plot(x, ysin_floor, dashes=[4, 5],label="floor")
p.plot(x, ysin_trunc, dashes=[4, 8],  label="trunc/fix")


p.grid(True, which='both')
p.axhline(y=0, color='k')
p.axvline(x=0, color='k')

p.legend(loc='lower right')

display(p)
```

### Minima/Maxima
>
> - np.**minimum**(x1,x2,...) [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.minimum.html) min value, elementwise, across 2 vectors, propagating nans
> - np.**maximum**(x1,x2,...) [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.maximum.html) max value, elementwise, across 2 vectors, propagating nans

> - np.**fmax**(x1,x2,...) [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.fmax.html) max value, elementwise, across 2 float vectors
> - np.**amin**(a, axis=None,...) [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.amin.html) min value across an axis
> - np.**amax**(a, axis=None,...) [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.amax.html) max value across an axis
> - np.**nanmin**(a, axis=None,...) [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.nanmin.html) min value, elementwise, across 2 vectors, ignoring nans

```python
# Fixing random state for reproducibility
np.random.seed(19680801)
N = 1000

def generate_stats(x, y):
    return {
        'xscatter': x,
        'yscatter': y,
        'xlinear': np.linspace(0,1.0,N),
        'stats': {
            'mean': np.repeat(np.mean(y),N),
            'mean': np.repeat(np.mean(y),N),
            'median': np.repeat(np.median(y),N),
            'std': np.repeat(np.std(y),N),
            'var': np.repeat(np.var(y),N),
            'percentile25' : np.repeat(np.percentile(y, 25),N),
            'percentile75' : np.repeat(np.percentile(y, 75),N)
        }
    }

def plot(p, s):
    p.scatter(s['xscatter'], s['yscatter'], s=4, alpha=0.4, label='items')
    for k, v in s['stats'].items():
        p.plot(s['xlinear'], v, label=k, dashes=[4,4])
    p.legend(loc='lower right');

configure_plot((0, 1.0), (0,1), (18, 6), 1, colors=['slategrey', 'r', 'orangered', 'goldenrod', 'y', 'forestgreen', 'dodgerblue' ])

random_stats = generate_stats( np.random.rand(N), np.random.rand(N) )
plt.subplot(1,2,1)
configure_subplot(plt)
plt.title('Random with Uniform Distribution')
plot(plt, random_stats)

normal_stats = generate_stats( np.random.normal(0.5, 0.15, size=N), np.random.normal(0.5, 0.15, size=N) )
plt.subplot(1,2,2)
configure_subplot(plt)
plt.title('Random with Normal Distribution')
plot(plt, normal_stats)

plt.show()
```

### Historgram
>
> - np.**histogram**(a, bins=10, range=None,...) [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.histogram.html) *creates histogram data for display*
> - np.**histogram2d**(x, y, bins=10, range=None,...) [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.histogram2d.html) *creates 2d histogram data for display*
> - np.**histogramdd**(sample, bins=10, range=None,...) [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.histogramdd.html) *creates multi-dimensional histogram data for display*
> - np.**bincount**(x, weights=None, minlength=0) [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.bincount.html) *returns frequency of each value*
> - np.**digitize**(x, bins, right=False) [➚](https://docs.scipy.org/doc/numpy/reference/generated/numpy.digitize.html) *returns bin index of which value fits in*

```python
# Fixing random state for reproducibility
#np.random.seed(19680801)

N = 4000
B = 50

configure_plot((0, 1.0), (0,1), (20, 6), 1)
data_uniform = np.random.rand(N)
data_normal = np.random.normal(0.5, 0.15, size=N)

m = plt.cm.get_cmap('magma')

plt.subplot(1,3,1)
configure_subplot(plt)
uy, ux = np.histogram(data_uniform, B, normed=True)
colors = [m(x) for x in ux]
plt.ylim(0.0, 4.0)
plt.bar(ux[:-1]*B+1, uy, align='center', alpha=0.5, color=colors)
plt.title('Histogram - Uniform Distribution')

plt.subplot(1,3,2)
configure_subplot(plt)
ny, nx = np.histogram(data_normal, B, normed=True)
plt.ylim(0.0, 4.0)
plt.bar(nx[:-1]*B+1, ny, align='center', alpha=0.5, color=colors)
plt.title('Histogram - Normal Distribution')
 
plt.subplot(1,3,3)
configure_subplot(plt)
xedges, yedges = np.linspace(0.0, 1.0, B), np.linspace(0.0, 1.0, B)
hist, xedges, yedges = np.histogram2d(data_uniform, data_normal, (xedges, yedges))
xidx = np.clip(np.digitize(data_uniform, xedges), 0, hist.shape[0]-1)
yidx = np.clip(np.digitize(data_normal, yedges), 0, hist.shape[1]-1)
c = hist[xidx, yidx]
plt.scatter(data_uniform, data_normal,s=1,  c=c, cmap='magma', alpha=0.5)
plt.title('Histogram2D - Uniform by Normal Distribution')

plt.show()
```
