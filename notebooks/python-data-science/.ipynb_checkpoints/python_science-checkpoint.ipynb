# Python - Scientific/Mathematical Computing

### Computation
- [NumPy](http://www.numpy.org/#) - Support Nth dimensional data structures - Vectors, Matrices, Tensors etc
- [SymPy](http://docs.sympy.org/latest/modules/evalf.html) - Computing algebraic system (CAS) - symbolic computing, no loss of accuracy, slower
- [SciPy](http://scipy.github.io/devdocs/index.html) - Advanced mathematical computation
- [GmPy](https://github.com/aleaxit/gmpy) (system module) - support infinitely long numberical values
